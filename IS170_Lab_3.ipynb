```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
#fix for an error with mlxtend package showing depreceation warnings
#run before running code below
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('products.csv')
df = df[['Products']]
df.head()
#reads products.csv and takes the products column to try and match what we see in the example
#must download csv from kaggle and use google colabs upload function to upload
```

```python
df.shape
#30,000 entries in products column
```

```python
data=list(df['Products'].apply(lambda x:x.split(",")))
data
#splits strings into a list using comma as delimiter
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)
df=df.astype(int)
df
#one hot encoding to label items that appear in each transaction as 1 and empty ones 0
```

```python
df=apriori(df,min_support=0.03,use_colnames=True)
#sets support value at 0.03 due to the large datset, support values are low
```

```python
df
```

```python
df_ar=association_rules(df,metric='confidence', min_threshold=0.2)
#sets the confidence threshold to 0.48; only associations with confidence > .48 are taken
```

```python
df_ar
#shows associations with a confidence level of 0.2 or above
```
